```python
europe = europe.loc[europe['teamId'] == 4418]
print(europe.to_string())
#print(europe['teamId'].dtype)
```

```python
firsthalf = europe.loc[europe['matchPeriod'] == '1H']
firsthalf
```

```python
secondhalf = europe.loc[europe['matchPeriod'] == '2H']
secondhalf
```

```python
firsthalf = firsthalf.loc[firsthalf['teamId'] == 4418]
firsthalf
```

```python
secondhalf = secondhalf.loc[secondhalf['teamId'] == 4418]
secondhalf
```

```python
x = []
y = []
for ind in firsthalf.index:
    x.append(firsthalf["positions"][ind][0]['x'])
    y.append(firsthalf["positions"][ind][0]['y'])
```

```python
plt.plot(x,y)
```

```python
x2 = []
y2 = []
for ind in secondhalf.index:
    x2.append(secondhalf["positions"][ind][0]['x'])
    y2.append(secondhalf["positions"][ind][0]['y'])
```

```python
plt.plot(x2,y2)
```

```python
plt.hist2d(x,y, bins=[np.arange(0,100,2),np.arange(0,100,2)])
plt.show()
```
